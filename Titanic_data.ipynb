```python
sns.boxplot(x='Pclass',y='Age',data=titanic_data)
```

```python
def age_handle(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
      return 29
    else:
      return 24
  else:
    return Age
```

```python
titanic_data['Age']=titanic_data[['Age','Pclass']].apply(age_handle,axis=1)
```
